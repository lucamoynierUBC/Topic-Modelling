```python
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_md")


def preprocessor(text_data):
    """
    given text, min token length, and irrelevant pos tags carry out preprocessing of the text and
    return a preprocessed string. 
    """
    preprocessed_text = []
    for data in text_data:
        preprocessed_text.append(data.get("text"))
    
    return list(nlp.pipe(preprocessed_text))

def tokenizer(doc, min_token_len, irrelevant_pos):
    text_df = {}
    for i, sentence in enumerate(doc):
        clean_text = []
        for token in sentence:
            if(
            token.is_stop == False # make sure token is not stopword
            and len(token) > min_token_len # make sure token is greater then specified length
            and token.pos_ not in irrelevant_pos
            ):
                lemma = token.lemma_  # Take the lemma of the word
                clean_text.append(lemma.lower())
        text_df[f"pp_text_{i+1}"] = " ".join(clean_text)
    return text_df
text_df = tokenizer(preprocessor(text_data), 2, ["ADV", "PRON", "CCONJ", "PUNCT", "PART", "DET", "ADP", "SPACE"])
df = pd.DataFrame.from_dict(text_df, orient='index', columns=['text_pp'])
df = df.reset_index(drop=True)
df
```
